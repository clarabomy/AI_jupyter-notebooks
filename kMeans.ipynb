# Lab Work 3 : Clustering with KMeans 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import sklearn.metrics as metrics

f = open("data/kmeans.csv")
# f.readline()  # if there is the header
data = np.loadtxt(f, delimiter='\t')
print(data)
```

```python
plt.scatter(data[:,0], data[:,1])
plt.axis([0, 100, 0, 100])
plt.show()
```

```python
euclidean_dists = metrics.euclidean_distances(data)

def plot_distance_matrix(D):
    fig, ax = plt.subplots()
    image = ax.matshow(D)
    fig.colorbar(image)
    ax.set_xticks([])
    ax.set_yticks([])
    fig.suptitle('Distance Matrix')
    
plot_distance_matrix(euclidean_dists)
```

```python
model = KMeans(n_clusters=3, random_state=10)
```

```python
clus = model.fit_predict(data)
print (clus)
```

```python
plt.scatter(data[:,0], data[:,1],  c=clus)

plt.axis([0, 100, 0, 100])
plt.show()
```

```python
plt.scatter(data[:,0], data[:,1],  c=clus)

centroids = model.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
plt.scatter(centroids_x, centroids_y, marker='X', s=200, c='r')

plt.axis([0, 100, 0, 100])
plt.show()
```

```python
markers = '^s+'
colors = 'rgb'
for i, m in enumerate(markers):
    points = (clus == i)
    c = [colors[l] for l in clus[points]]
    plt.scatter(data[points,0], data[points,1], c=c, marker=m)
plt.axis([0, 100, 0, 100])

plt.show()
```

```python
idx = np.argsort(clus)
rearranged_dists = euclidean_dists[idx,:][:,idx]
plot_distance_matrix(rearranged_dists)
```
