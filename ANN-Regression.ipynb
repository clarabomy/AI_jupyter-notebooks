```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
f = open("data/mydata.csv")

f.readline()  # skip the header line (X,Y)
data = np.loadtxt(f, delimiter=',')

print(data)
```

```python
plt.scatter(data[:,0], data[:,1])
plt.axis([0, 100, 0, 100])
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
X = data[:,0]
Y = data[:,1]
```

```python
model = Sequential()
model.add(Dense(5, input_dim=1, kernel_initializer='normal', activation='sigmoid'))
model.add(Dense(5, input_dim=1, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, activation='relu'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
```

```python
model.fit(X, Y, epochs=2000, verbose=2)
```

```python
Y_pred = model.predict(np.array(X))
plt.scatter(X, Y,  color='black')
plt.plot(X, Y_pred, color='red', linewidth=3)

plt.show()
```
